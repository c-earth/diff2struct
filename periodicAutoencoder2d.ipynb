# Generate 2D periodic data

```python
import numpy as np
```

```python
data_num = 4096
n_max = 25
resolution = 64
```

```python
values = np.random.choice(2, (data_num, n_max))
locations = np.random.choice(resolution, (data_num, n_max, 2))
```

```python
data = np.zeros((data_num, resolution, resolution))
```

```python
data[np.broadcast_to(np.arange(data_num)[:, None], (data_num, n_max)), locations[:, :, 0], locations[:, :, 1]] = values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.imshow(data[0])
plt.show()
```
